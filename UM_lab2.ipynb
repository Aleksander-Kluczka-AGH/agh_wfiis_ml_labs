```python
# zad1
x = np.arange(0, 10, 1)
y = np.exp(-x)
x_err = np.sqrt(y) * 0.3
y_err = 0.1
plt.errorbar(x, y, x_err, y_err, marker="o", label="one")
plt.errorbar(
    x,
    y - 1,
    x_err * 0.3,
    y_err + 0.5,
    marker=4,
    label="two",
    xuplims=[True, False] * 5,
    xlolims=[False, True] * 5,
)
plt.errorbar(x, y - 2, x_err, y_err, marker="s", label="three")
plt.errorbar(x, y - 3, x_err, y_err, marker="3", label="four")
plt.legend()
plt.show()
```

```python
# zad2
x_data = []
y_data = []
for i in range(13):
    x_data.append(f"{2000+i}-{1+i}-{15+i}")
    y_data.append(1.3 + 0.79 * i**3 - np.exp(0.3 * i + np.sqrt(i)))

plt.figure(figsize=(15, 5))
plt.plot(x_data, y_data, marker="o")
plt.title("Dates")
plt.xlabel("date")
plt.ylabel("mah weight")
plt.grid()
```

```python
# zad3
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

planets = sns.load_dataset("planets")
print(planets)
```

```python
g = sns.relplot(data=planets, x="distance", y="orbital_period")
```

```python
g = sns.relplot(data=planets, x="distance", y="orbital_period", hue="year", size="mass")
```

```python
sns.set_theme(style="whitegrid")
cmap = sns.cubehelix_palette(rot=0.2, as_cmap=True)

g = sns.relplot(
    data=planets,
    x="distance",
    y="orbital_period",
    hue="year",
    size="mass",
    palette=cmap,
    sizes=(10, 200),
)

g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=0.25)
g.ax.yaxis.grid(True, "minor", linewidth=0.25)
g.despine(left=True, bottom=True)
g.fig.set_figwidth(12.27)
g.fig.set_figwidth(7.7)
```

```python
tips = sns.load_dataset("tips")

tips[0:9]
```

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size"
)
```

```python
g = sns.FacetGrid(data=tips, col="smoker", row="sex")
g.map(sns.regplot, "total_bill", "tip")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
ax1 = sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", ax=ax1)
ax2 = sns.swarmplot(data=tips, x="day", y="tip", hue="sex", ax=ax2)
```

```python
sns.displot(data=tips, x="total_bill")
sns.displot(data=tips, x="total_bill", hue="sex")
```
